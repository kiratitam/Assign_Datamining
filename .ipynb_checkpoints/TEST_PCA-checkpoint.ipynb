```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from tqdm import tqdm
from sklearn.decomposition import PCA

DATADIR = "C:\\Users\\Luxuss\\Desktop\\SharedFolder\\Assign_Datamining\\Original\\data"

CATEGORIES = ["Aston Martin V8 Vantage Convertible 2012", "Audi S4 Sedan 2007", "Bentley Continental Flying Spur Sedan 2007", \
              "BMW M3 Coupe 2012", "Chevrolet Corvette ZR1 2012", "Ford GT Coupe 2006", "Jaguar XK XKR 2012", \
              "Lamborghini Diablo Coupe 2001", "Mercedes-Benz S-Class Sedan 2012", "Rolls-Royce Phantom Sedan 2012"]

for category in CATEGORIES:  # do dogs and cats
    path = os.path.join(DATADIR,category)  # create path to dogs and cats
    for img in os.listdir(path):  # iterate over each image per dogs and cats
        img_array = cv2.imread(os.path.join(path,img) ,cv2.IMREAD_GRAYSCALE)  # convert to array
        plt.imshow(img_array, cmap='gray')  # graph it
        plt.show()  # display!
        break
    break
```

```python
IMG_SIZE = 150

new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
new_array
```

```python
new_array.shape
```

```python
pca = PCA(0.95)
```

```python
data = pca.fit_transform(new_array)
data.shape
```

```python
result = pca.inverse_transform(data)
```

```python
plt.figure(figsize=(8,4))
#image feature 784
plt.subplot(1,2,1)
plt.imshow(new_array,cmap=plt.cm.gray,interpolation="nearest")
plt.title("Original")
#image feature 95% -> 154 
plt.subplot(1,2,2)
plt.imshow(result,cmap=plt.cm.gray,interpolation="nearest")
plt.title("PCA")
plt.show()
```

```python
training_data = []
tmp = 1
def preprocess_data_pca(tmp):
    for category in CATEGORIES:
        path = os.path.join(DATADIR,category)
        class_num = CATEGORIES.index(category)
        for img in tqdm(os.listdir(path)):
            try:
                img_path = "C:\\Users\\Luxuss\\Desktop\\SharedFolder\\Assign_Datamining\\Pre_Process_PCA\\data\\{0}\\{1}.jpg"\
                .format(category, tmp)
                tmp += 1
                img_array = cv2.imread(os.path.join(path,img) ,cv2.IMREAD_GRAYSCALE)
                new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
                data = pca.fit_transform(new_array)
                result = pca.inverse_transform(data)
                cv2.imwrite(img_path, result)
            except Exception as e:
                print(e)

preprocess_data_pca(tmp)
```
